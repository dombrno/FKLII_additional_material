```python
def get_binding_energy(eff_mass, epsilon):
    return (eff_mass / epsilon**2 * (cst.e**4 * cst.m_e / (2.0 * cst.hbar**2))/
            (4.0 * math.pi * cst.epsilon_0)**2 / cst.e * 1000.0)
```

```python
print ("GaAs binding energy", get_binding_energy(0.07, 12.5), "meV")
```

```python
print ("InSb binding energy", get_binding_energy(0.014, 16.8), "meV")
```
